```python
from diffusers import DiffusionPipeline
from PIL import Image
import numpy as np
```

```python
pipeline = DiffusionPipeline.from_pretrained("Qilex/VirtualPet-Gen")

def generate_pets(num_to_generate):
    #generate list of num_to_generate pets
    images = pipeline(num_to_generate)["sample"]
    return images

def concatenate_imgs(imgs):
    #concatenate 1-4 images into a grid
    length = len(imgs)
    if length == 1:
        return imgs[0]
    top = Image.fromarray(np.concatenate([np.array(x) for x in imgs[:2]],axis=1))
    if length == 2:
        return top
    if len(imgs)==3:
        fake = np.zeros([128,128,3],dtype=np.uint8)
        fake[:] = 255
        bottom = Image.fromarray(np.concatenate([imgs[2], fake],axis=1))
    elif len(imgs)==4:
        bottom = Image.fromarray(np.concatenate([imgs[2], imgs[3]],axis=1))
    return Image.fromarray(np.concatenate([top,bottom],axis=0))
```

```python
#generate a list of image objects
imgs = generate_pets(4)

#concatenate 4 images into a grid
image_grid = concatenate_imgs(imgs)

#display
display(image_grid)
```
